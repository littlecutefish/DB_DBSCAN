```python
import seaborn as sns;sns.set()

from sklearn.datasets import load_digits
%matplotlib inline
```

```python
dx = load_digits()   #读取数字数据，介绍如下
''' 
    Each datapoint is a 8x8 image of a digit.
    
    =================   ==============
    Classes                         10
    Samples per class             ~180
    Samples total                 1797
    Dimensionality                  64
    Features             integers 0-16
    =================   ==============
'''
xx = dx.data         #数据
yy = dx.target       #标签
d = 3                #设置需要降到的维数
```

## 降维
PCA, KernelPCA, NMF, FastICA, TruncatedSVD, SparsePCA, IncrementalPCA, DictionaryLearning, MiniBatchNMF, LDA (Latent Dirichlet Allocation)

```python
from sklearn.decomposition import PCA, KernelPCA, FastICA, NMF, SparsePCA, IncrementalPCA, DictionaryLearning, MiniBatchNMF
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import TruncatedSVD

# 假設 xx 和 yy 是你的數據集和標籤
# xx: 特徵矩陣 (例如：1797 x 特徵維度)
# yy: 標籤向量

# 設定 d (目標維度)
d = 3  # 例如降到3維

# 1. PCA (Principal Component Analysis)
pca = PCA(n_components=d)
pca_result = pca.fit_transform(xx)
pca_result = MinMaxScaler().fit_transform(pca_result)

# 2. KernelPCA (使用RBF核)
kernel_pca = KernelPCA(n_components=d, kernel='rbf')
kernel_pca_result = kernel_pca.fit_transform(xx)
kernel_pca_result = MinMaxScaler().fit_transform(kernel_pca_result)

# 3. FastICA (獨立成分分析)
fast_ica = FastICA(n_components=d)
fast_ica_result = fast_ica.fit_transform(xx)
fast_ica_result = MinMaxScaler().fit_transform(fast_ica_result)

# 4. LDA (線性判別分析)
lda = LDA(n_components=d)
lda_result = lda.fit_transform(xx, yy)
lda_result = MinMaxScaler().fit_transform(lda_result)

# 5. LLE (局部線性嵌入)
from sklearn.manifold import LocallyLinearEmbedding
lle = LocallyLinearEmbedding(n_components=d)
lle_result = lle.fit_transform(xx)
lle_result = MinMaxScaler().fit_transform(lle_result)

# 6. t-SNE (t-分佈隨機鄰域嵌入)
tsne = TSNE(n_components=d)
tsne_result = tsne.fit_transform(xx)
tsne_result = MinMaxScaler().fit_transform(tsne_result)

# 7. NMF (非負矩陣分解)
nmf = NMF(n_components=d)
nmf_result = nmf.fit_transform(xx)
nmf_result = MinMaxScaler().fit_transform(nmf_result)

# 8. TruncatedSVD (截斷奇異值分解)
svd = TruncatedSVD(n_components=d)
svd_result = svd.fit_transform(xx)
svd_result = MinMaxScaler().fit_transform(svd_result)

# 9. SparsePCA (稀疏主成分分析)
sparse_pca = SparsePCA(n_components=d)
sparse_pca_result = sparse_pca.fit_transform(xx)
sparse_pca_result = MinMaxScaler().fit_transform(sparse_pca_result)

# 10. IncrementalPCA (增量主成分分析)
incremental_pca = IncrementalPCA(n_components=d)
incremental_pca_result = incremental_pca.fit_transform(xx)
incremental_pca_result = MinMaxScaler().fit_transform(incremental_pca_result)

# 11. DictionaryLearning (字典學習)
dictionary_learning = DictionaryLearning(n_components=d)
dictionary_learning_result = dictionary_learning.fit_transform(xx)
dictionary_learning_result = MinMaxScaler().fit_transform(dictionary_learning_result)

# 12. MiniBatchNMF (小批量NMF)
mini_batch_nmf = MiniBatchNMF(n_components=d)
mini_batch_nmf_result = mini_batch_nmf.fit_transform(xx)
mini_batch_nmf_result = MinMaxScaler().fit_transform(mini_batch_nmf_result)
```

## 聚類：DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from munkres import Munkres

# DBSCAN 初始化
db = DBSCAN(eps=0.07, min_samples=50)

# 混淆矩陣繪圖
def draw_confusion_matrix(label, predict):
    sns.heatmap(confusion_matrix(label, predict).T, square=True, annot=True, fmt='d', cbar=False, cmap='rainbow')
    plt.xlabel('True Label')
    plt.ylabel('Predicted Label')

# 定義對齊函數
def predict_transfer(label, predict):
    L1 = np.array(label)
    L2 = np.array(predict)
    Label1 = np.unique(L1)
    nClass1 = len(Label1)
    Label2 = np.unique(L2)
    nClass2 = len(Label2)
    nClass = np.maximum(nClass1, nClass2)
    G = np.zeros((nClass, nClass))
    
    for i in range(nClass1):
        ind_cla1 = L1 == Label1[i]
        ind_cla1 = ind_cla1.astype(float)
        for j in range(nClass2):
            ind_cla2 = L2 == Label2[j]
            ind_cla2 = ind_cla2.astype(float)
            G[i, j] = np.sum(ind_cla2 * ind_cla1)
    
    m = Munkres()
    index = m.compute(-G.T)
    index = np.array(index)
    c = index[:, 1]
    newL2 = np.zeros(L2.shape)
    
    for i in range(nClass2):
        newL2[L2 == Label2[i]] = Label1[c[i]]
    
    return newL2

# 聚類和準確度計算
def dbscan_accuracy_method(data, label, method_name):
    # 聚類預測
    y_pred = db.fit_predict(data)
    y_pred = predict_transfer(label, y_pred)  # 對齊預測標籤與真實標籤
    accuracy = accuracy_score(label, y_pred)  # 計算準確度
    print(f"{method_name} 準確度: {accuracy:.4f}")
    return accuracy, y_pred  # 返回準確度和預測結果

# 存儲每個方法的準確度和預測結果
accuracies = []
predictions = []

# 12種降維方法
methods = [
    (pca_result, 'PCA'),
    (kernel_pca_result, 'KernelPCA'),
    (nmf_result, 'NMF'), 
    (fast_ica_result, 'FastICA'),
    (svd_result, 'TruncatedSVD'),
    (sparse_pca_result, 'SparsePCA'),
    (incremental_pca_result, 'IncrementalPCA'),
    (dictionary_learning_result, 'DictionaryLearning'),
    (mini_batch_nmf_result, 'MiniBatchNMF'),
    (lda_result, 'LDA'),
    (lle_result, 'LLE'),
    (tsne_result, 't-SNE'),
]

# 對每個降維方法進行DBSCAN聚類並計算準確度
for data, method_name in methods:
    accuracy, y_pred = dbscan_accuracy_method(data, yy, method_name)
    accuracies.append(accuracy)  # 將每個方法的準確度加入列表
    predictions.append((method_name, y_pred))  # 將每個方法的預測結果加入列表

# 計算結果
methods_labels = ['PCA', 'KernelPCA', 'FastICA', 'LDA', 'LLE', 't-SNE', 
                  'NMF', 'TruncatedSVD', 'SparsePCA', 'IncrementalPCA', 
                  'DictionaryLearning', 'MiniBatchNMF']

# 繪製12個混淆矩陣
figure = plt.figure(figsize=(18, 12))

# 使用for迴圈動態創建每個子圖
for i, (method_name, y_pred) in enumerate(predictions):
    ax = figure.add_subplot(3, 4, i+1, aspect='equal')
    draw_confusion_matrix(yy, y_pred)  # 繪製混淆矩陣
    ax.set_title(method_name)  # 設定每個子圖的標題

# 設定主標題
figure.suptitle('DBSCAN', fontsize=22)

# 顯示圖形
plt.tight_layout()
plt.subplots_adjust(top=0.9)  # 調整主標題的高度
plt.show()

# 繪製條形圖來顯示準確度比較
size = len(accuracies)  # 使用準確度的數量
x = np.arange(size)

# 設定條形圖的寬度
total_width, n = 0.6, 1  # 只有一個條形
width = total_width / n
x = x - (total_width - width) / 2  # 使條形居中

# 繪製條形圖
plt.bar(x, accuracies, width=width, label='Accuracy')

# 設定標籤和標題
plt.xticks(x, methods_labels, rotation=45, ha='right')  # 橫坐標標籤
plt.xlabel('Dimensionality Reduction Methods')
plt.ylabel('Accuracy')
plt.title('DBSCAN Clustering Accuracy for Different Dimensionality Reduction Techniques')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.tight_layout()
plt.show()
```
